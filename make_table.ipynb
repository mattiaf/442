```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt

#
# --- Seasons to load --- 
# Files have been pickled with the scrape.ipynb file, which scrapes different seasons in the football-lineups.com website
#

folder = 'Seasons/'
seasons=[ "FA_PL2003-2004.pickle","FA_PL2005-2006.pickle","FA_PL2007-2008.pickle","FA_PL2009-2010.pickle","FA_PL2011-2012.pickle", "FA_PL2013-2014.pickle"]
years =['2004','2006','2008','2010','2012', '2014']

#
# --- Join data ----
# Creates a Data Frame Including all the past seasons. 
# Each game played by a team in one year is represented by a different row
#

df_all = pd.DataFrame(data=None) 
for i in range(0,len(seasons)):
    df=pickle.load( open(folder+seasons[i], "rb" ) )
    df_toadd= pd.DataFrame(data=None) 
    df_toadd['Tactic']=pd.Series(df.Tactic1.append(df.Tactic2)) # append home games and away games
    df_toadd['Team']=pd.Series(df.Team1.append(df.Team2)) # append home games and away games
    df_toadd['Year'] = [years[i]]*len(df_toadd) # append year
    df_all = df_all.append(df_toadd)

df_all.head()
```

```python
#
# Exploration: which lineups has Liverpool used in 2008?
#

df_all[(df_all.Team == 'Liverpool') & (df_all.Year == '2008')].groupby('Tactic').count()['Team'].plot(kind='bar', alpha=0.8, color='red')
```

```python
#
# -- How many times has a tactic been used in a year? --
# Let's 'pivot' our dataframe
# Rows will be years, columns different tactics, and values number of times the tactic was used

df_tactics=df_all.groupby(['Year','Tactic']).agg(lambda x: sum(x.value_counts()))
df_tactics['Year']=df_tactics.index.get_level_values('Year')
df_tactics['Tactic']=df_tactics.index.get_level_values('Tactic')
df_tactics['Year'] = df_tactics['Year'].apply(lambda x: str(int(x)-1) + '-' + x)

df_pivot=df_tactics.pivot(index = 'Year', columns='Tactic', values='Team')
df_pivot=df_pivot.fillna(0)
df_pivot.head()

```

```python
#
# Exploration: how many times has been 4-4-2 used? and 4-2-3-1?
#


df_pivot['4-4-2'].plot()
df_pivot['4-2-3-1'].plot()
```

```python
# plt.show()
df_pivot.columns=['tactic'+i.replace('-','') for i in df_pivot.columns]
mintimes=100
df_pivot['Others']=df_pivot[df_pivot.columns[df_pivot.sum() < mintimes]].sum(axis=1)
df_pivot[df_pivot.columns[df_pivot.sum() > mintimes]].to_csv('tacticsPL.csv')
```
